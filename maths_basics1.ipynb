```python
#math python library

#import library

import math

#exponents

#2 to the power raised to 3

print(2**3)
#base = 2 power = 3
print(5**3)

#square loop

print(math.sqrt(25))
#sqrt is name of function
#25 is the parameter of that you are passing to the function
#math is where the function is located


#power function
print(math.pow(2, 3))
#2 is base
#3 is power
print(math.pow(-2, 5))

print("exponential of 2:", math.exp(1))


print("natural log of 10:", math.log(10))
#log(10) or ln(10)
#log base is e, e=2.8

#log is how many times divide number by itself to get to a certain number

print("log base 10 of 100", math.log(100))

#absolute value
print("absolute value of -5", math.fabs(-5))
#takes away start

#factorial
print("factorial of 5", math.factorial(5))


#convert radians to degrees
angle = 45 #degrees
radians = math.radians(angle)
print(radians)

#whenever you want to use trigonometric function you have to convert the angle to radians (sin, cos, tan)

print("sine of 45 degrees",math.sin(radians))

print("cos of 45 degrees", math.cos(radians))

print("tan of 45 degrees", math.tan(radians))

#inverse trigo
print("arcsin of 8.5", math.asin(0.7071067811865476))

print("arccos of 8.5", math.asin(0.7071067811865476))


angle1 = 60
radians1 = math.radians(angle1)

print(radians1)

print("sine of 60 degrees",math.sin(radians1))
print("cos of 60 degrees",math.cos(radians1))
print("tan of 60 degrees",math.tan(radians1))

print("arcsin of 0.86...", math.asin(0.866025403784438))
print("arcos of 0.5", math.acos(0.5000000000000001))
print("arctan of 1.73", math.atan(1.7320508075688767))

print("degrees arcsin =", math.degrees(1.0471975511965965))
print("degrees arcos =", math.degrees(1.0471975511965976))
print("degrees atan =", math.degrees(1.0471975511965976))

#mathematical constants
print("value of pi:", math.pi)
print("value of e:", math.e)
print("value of tau:", math.tau) # 2*pi
print("value of infinity:", math.inf)
print("value of not-a-number:", math.nan)
```
